```python
import yfinance as yf
import mplfinance as mpf
import numpy as np

def plot_stock_chart_with_signals(stock_symbol):
    # Download historical stock price data from Yahoo Finance
    data = yf.download(stock_symbol, start="2021-01-01", end="2021-09-01")

    # Calculate buy and sell signals based on MA crossovers
    data['MA20'] = data['Adj Close'].rolling(20).mean()
    data['MA50'] = data['Adj Close'].rolling(50).mean()
    data = data.dropna()

    buy_signals = []
    sell_signals = []

    for i in range(len(data)):
        if (data['MA20'].iloc[i] > data['MA50'].iloc[i]) and (data['MA20'].iloc[i-1] < data['MA50'].iloc[i-1]):
            buy_signals.append(data.iloc[i]['Adj Close'] * 0.98)
        else:
            buy_signals.append(np.nan)

        if (data['MA20'].iloc[i] < data['MA50'].iloc[i]) and (data['MA20'].iloc[i-1] > data['MA50'].iloc[i-1]):
            sell_signals.append(data.iloc[i]['Adj Close'] * 1.02)
        else:
            sell_signals.append(np.nan)

    # Create buy and sell markers for the candlestick chart
    buy_markers = mpf.make_addplot(buy_signals, type='scatter', markersize=120, marker='^', color='g', secondary_y=False)
    sell_markers = mpf.make_addplot(sell_signals, type='scatter', markersize=120, marker='v', color='r', secondary_y=False)
    apds = [buy_markers, sell_markers]

    # Define custom styling for the candlestick chart
    custom_style = mpf.make_mpf_style(base_mpf_style='charles', gridstyle=':', gridcolor='gray', rc={'axes.labelcolor': 'gray'})

    # Create the candlestick chart with buy/sell signals and volume bars
    mpf.plot(data, type="candle", addplot=apds, title=f"{stock_symbol} Candlestick Chart with Buy/Sell Signals", style=custom_style, volume=True, figratio=(10, 6))

# Input the stock symbol from the user
stock_symbol = input("Enter a stock symbol (e.g., AAPL, MSFT, GOOG): ")

# Call the function to plot the stock chart with buy/sell signals
plot_stock_chart_with_signals(stock_symbol)
```
